```python
import pandas as pd
```

### Loading Data

Combination of Deaths and Cases

```python
df_cases = pd.read_csv('../data/epidemic/cases_state.csv')
df_deaths = pd.read_csv('../data/epidemic/deaths_state.csv')
```

```python
df_cases.head()
```

```python
df_deaths.head()
```

```python
len(df_cases)
```

```python
df_casedeath = pd.merge(df_cases,df_deaths,how='left',on=['date','state'])
df_casedeath
```

```python
df_casedeath.describe()
```

```python
df_casedeath.to_csv("../consolidated_data/state_cases_death.csv")
```

Healthcare Facilities


```python
df_hospital= pd.read_csv('../data/epidemic/hospital.csv')
df_icu = pd.read_csv('../data/epidemic/icu.csv')
df_pkrc = pd.read_csv('../data/epidemic/pkrc.csv')
```

```python
for i in df_hospital.columns[2::]:
    df_hospital.rename(columns={i:i+"_hospital"},inplace=True)
```

```python
for i in df_pkrc.columns[2::]:
    df_pkrc.rename(columns={i:i+"_pkrc"},inplace=True)
```

```python
df_hospital.head(3)
```

```python
df_icu.head(3)
```

```python
df_pkrc.head(3)
```

```python
df_icu_hospital = pd.merge(df_hospital,df_icu,how='left',on=['date','state'])
df_icu_hospital.describe()
```

```python
df_icu_hospital["state"].unique()
```

```python
df_pkrc["state"].unique()
```

```python
if(len(df_icu_hospital["state"].unique()) == len(df_pkrc["state"].unique())):
    print("Yes")
else:
    print("No")
```

```python
df_healthcare_facilities = pd.merge(df_icu_hospital,df_pkrc,how='left',on=['date','state'])
```

```python
df_healthcare_facilities.columns
```

```python
df_healthcare_facilities.to_csv("../data/consolidated_data/healthcare_facilities.csv")
```

All state related sets

```python
df_vax = pd.read_csv("../data/vaccination/vax_state.csv")
df_tests = pd.read_csv("../data/epidemic/tests_state.csv")
```

```python
df_main = pd.merge(df_casedeath,df_tests,how='left',on=['date','state'])
df_main = pd.merge(df_main,df_healthcare_facilities,how='left',on=['date','state'])
df_main = pd.merge(df_main,df_tests,how='left',on=['date','state'])
df_main.info()
```

```python
df_vax_state = pd.read_csv('../data/vaccination/vax_state.csv')
df_main = pd.read_csv('../data/consolidated_data/all_state.csv')
df_main = pd.merge(df_main,df_vax_state,how='left',on=['date','state'])
```

```python
df_main.describe()
```

```python
df_main.to_csv("../data/consolidated_data/all.csv")
```
