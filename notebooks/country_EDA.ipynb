### Data Breakdown

### Cases, Testing and Deaths

1) `date`: yyyy-mm-dd format; data correct as of 1200hrs on that date
2) `cases_new`: cases reported in the 24h since the last report
3) `cases_import`: imported cases reported in the 24h since the last report
4) `cases_active`: Covid+ individuals who have not recovered or died
5) `cases_recovered` recovered cases reported in the 24h since the last report
6) `cases_cluster`: number of cases attributable to clusters; the difference between `cases_new` and the sum of cases attributable to clusters is the number of sporadic cases
7) `cluster_x`: cases attributable to clusters under category `x`; possible values for `x` are import, religious, community, highRisk, education, detentionCentre, and workplace
8) `cases_agecat`: cases falling into one of 4 age categories, i.e. child (0-11), adolescent (12-17), adult (18-59), elderly (60+); note that the sum of cases by age may not equal the total cases for that day, as some cases are registered without ages or with unverifiable age data
9) `cases_pvax`: number of partially-vaccinated individuals who tested positive for Covid (perfect subset of `cases_new`), where "partially vaccinated" is defined as receiving at least 1 dose of a 2-dose vaccine at least 1 day prior to testing positive, or receiving the Cansino vaccine between 1-27 days before testing positive
10) `cases_fvax`: number of fully-vaccinated who tested positive for Covid (perfect subset of `cases_new`), where "fully vaccinated" is defined as receiving the 2nd dose of a 2-dose vaccine at least 14 days prior to testing positive, or receiving the Cansino vaccine at least 28 days before testing positive
11) `rtk-ag`: number of tests done using Antigen Rapid Test Kits (RTK-Ag)
12) `pcr`: number of tests done using Real-time Reverse Transcription Polymerase Chain Reaction (RT-PCR) technology
13) `deaths_new`: deaths due to COVID-19 based on **date reported to public**
14) `deaths_bid`: deaths due to COVID-19 which were brought-in dead based on **date reported to public** (perfect subset of `deaths_new`)
15) `deaths_new_dod`: deaths due to COVID-19 based on **date of death**
16) `deaths_bid_dod`: deaths due to COVID-19 which were brought-in dead based on **date of death** (perfect subset of `deaths_new_dod`)
17) `deaths_pvax`: number of partially-vaccinated individuals who died due to COVID-19 based on **date of death** (perfect subset of 18)`deaths_new_dod`: where "partially vaccinated" is defined as receiving at least 1 dose of a 2-dose vaccine at least 1 day prior to testing positive, or receiving the Cansino vaccine between 1-27 days before testing positive.
19) `deaths_fvax`: number of fully-vaccinated who died due to COVID-19 based on **date of death** (perfect subset of `deaths_new_dod`), where "fully vaccinated" is defined as receiving the 2nd dose of a 2-dose vaccine at least 14 days prior to testing positive, or receiving the Cansino vaccine at least 28 days before testing positive.
20) `deaths_tat`: median days between date of death and date of report for all deaths reported on the day

### IP (Initial Preprocessing)

Dropping Columns based on Documentation

```python
df.drop(columns=['cases_cluster','deaths_new','deaths_bid','deaths_tat'],inplace=True)
```

These columns are dropped as :
* cases_cluster : Has already more grnaular variables available.
* deaths_new & deaths_bid : These are reported to public dates, not actual date of deaths.
* deaths_tat : Represents the median days of reporting delay of the death to the public

Imputing Missing Values

```python
df[['cluster_import', 'cluster_religious', 'cluster_community', 'cluster_highRisk',
    'cluster_education', 'cluster_detentionCentre', 'cluster_workplace']] = df[['cluster_import',
       'cluster_religious', 'cluster_community', 'cluster_highRisk',
       'cluster_education', 'cluster_detentionCentre', 'cluster_workplace']].fillna(0)
#Imputed with 0 to mimic to "no cases reported"
```

```python
df[['deaths_new_dod', 'deaths_bid_dod',
       'deaths_unvax', 'deaths_pvax', 'deaths_fvax', 'deaths_boost']] =df[['deaths_new_dod', 'deaths_bid_dod',
       'deaths_unvax', 'deaths_pvax', 'deaths_fvax', 'deaths_boost']].fillna(0)
#Imputed with 0 to mimic to "no deaths reported"
```

```python
df['rtk-ag'] = df['rtk-ag'].fillna(df['rtk-ag'].mean())
df['pcr'] = df['pcr'].fillna(df['pcr'].mean())
#Impute testing with mean, as NA does not mean 0 tests were done on that particular day.
```
