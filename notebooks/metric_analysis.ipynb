```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../data/consolidated_data/model_metrics.csv')
```

```python
df
```

```python
df_sorted = df.sort_values(by='Mean Square Error')
```

```python
plt.figure(figsize=(7,3))
sns.barplot(y=df['Model Name'],x=df['Mean Square Error'],color='red')
plt.title('Model Comparison by Mean Square Error')
plt.legend()
plt.show()
```

```python
df_sorted = df.sort_values(by='Root Mean Square Error')
plt.figure(figsize=(7,3))
sns.barplot(y=df_sorted['Model Name'],x=df_sorted['Root Mean Square Error'],color='orange')
plt.title('Model Comparison by Root Mean Square Error')
plt.legend()
plt.show()
```

```python
df_sorted = df.sort_values(by='Mean Absolute Error')
plt.figure(figsize=(7,3))
sns.barplot(y=df_sorted['Model Name'],x=df_sorted['Mean Absolute Error'],color='purple')
plt.title('Model Comparison by Mean Absolute Error')
plt.legend()
plt.show()
```

```python
df_sorted = df.sort_values(by='R2',ascending=False)
plt.figure(figsize=(7,3))
sns.barplot(y=df_sorted['Model Name'],x=df_sorted['R2'],color='cyan')
plt.title('Model Comparison by R2 Score')
plt.legend()
plt.show()
```
